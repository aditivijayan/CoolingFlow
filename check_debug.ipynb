```python
# data = np.loadtxt("flow_debug.txt")
file_list = ["flow_debug_0.8kpc_sol.txt", "flow_debug_0.8kpc_nosol.txt"]
label = ['solved', 'not solved']
i=1
plt.figure(figsize=(12,8))
for file in file_list[1:]:
    data = np.loadtxt(file)
    r_kpc        = data[:, 0]
    dlnrho_dlnr  = data[:, 1]
    plt.plot(r_kpc, dlnrho_dlnr, label=label[i])
    print(r_kpc.shape)
    i+=1
plt.legend()
plt.yscale('symlog', linthresh=1.e-9)
plt.xscale('log')
plt.xlim(0.7, 1.e2)
# plt.xlim(0.62, 0.64)
# dlnT_dlnr    = data[:, 2]
# Mach         = data[:, 3]
# v= data[:, 4]
# cs= data[:, 5]
# rho= data[:, 6]
# T= data[:, 7]

```

```python

file_list = ["flow_debug_0.8kpc_sol.txt", "flow_debug_0.8kpc_nosol.txt"]
label = ['solved', 'not solved']
i=1
for file in file_list[1:]:
    data = np.loadtxt(file)
    r_kpc        = data[:, 0]
    rho= data[:, 6]
    T= data[:, 7]
    # mask = (r_kpc>0.0) & (r_kpc<1.)
    plt.plot(r_kpc, T/1.e6, ls='-', label=label[i])
    plt.plot(r_kpc, rho, ls='-')
    print(np.amax(r_kpc))
    i+=1
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.xlim(0.7, 1.e2)
# plt.xlim(0.6, 0.7)
# dlnT_dlnr    = data[:, 2]
# Mach         = data[:, 3]
# v= data[:, 4]
# cs= data[:, 5]
# rho= data[:, 6]
# T= data[:, 7]

```
